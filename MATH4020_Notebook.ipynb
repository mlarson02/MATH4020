```python
import pandas as pd
import numpy as np

def append_column_of_ones(csv_file_path):
    # Read CSV file into a Pandas DataFrame
    dataframe = pd.read_csv(csv_file_path)

    # Append a column of '1's to the DataFrame
    dataframe['label'] = 1

    return dataframe

def append_column_of_zeros(csv_file_path):
    # Read CSV file into a Pandas DataFrame
    dataframe = pd.read_csv(csv_file_path)

    # Append a column of '1's to the DataFrame
    dataframe['label'] = 0

    return dataframe
```

```python
def filter_dataframe(df):
    """
    Filter DataFrame based on specified conditions.

    Parameters:
        df (pd.DataFrame): The DataFrame to filter.

    Returns:
        pd.DataFrame: Filtered DataFrame containing rows that pass all conditions.
    """
    # Define conditionals based on column values
    conditional1 = (df['n_RecoElectrons'] >= 2) | (df['n_RecoMuons'] >= 2)
    conditional2 = (df['Reco_ee_invMass'].between(70, 110)) | (df['Reco_mumu_invMass'].between(70, 110))

    # Combine conditionals using logical AND
    combined_mask = conditional1 & conditional2

    # Filter DataFrame based on combined mask
    filtered_df = df[combined_mask]
   
    indices_to_remove = list(range(33, 51))
    df_filtered2 = filtered_df.drop(filtered_df.columns[indices_to_remove], axis=1)
    indices_to_remove2 = list(range(3,23)) # remove jet params 
    df_filtered3 = df_filtered2.drop(df_filtered2.columns[indices_to_remove2], axis=1)
    
    return df_filtered3
```

```python
def calculate_max(entry):
    # Check if entry is a string
    if isinstance(entry, str):
        # Remove leading and trailing brackets, split the string by spaces, and convert to list of floats
        entry_list = [float(value) for value in entry.strip('[]').split()]
        # Calculate the average if the list is not empty
        if entry_list:
            return max(entry_list)#sum(entry_list) / len(entry_list)
        else:
            return 0  # Return '0' if the list is empty
    elif isinstance(entry, list):
        # Calculate the average of the list
        if entry:
            return max(entry_list)#sum(entry) / len(entry)
        else:
            return 0  # Return '0' if the list is empty
    else:
        return entry  # Return the entry unchanged if it's not a list or string
```

```python
def avg_columns(df, column_indices):
    df.iloc[:, column_indices] = df.iloc[:, column_indices].map(calculate_max)
    df = df.drop(columns=['RecoParticleTrack_absZ0sig'])
    df = df.drop(columns=['n_seltracks_DVs'])
    for column in df.columns:
        if df[column].dtype == 'object':
            try:
                df[column] = pd.to_numeric(df[column], errors='coerce')
            except ValueError:
                print(f"Unable to convert column '{column}' to numeric.")
    return df
```

```python
def avg_columns2(df, column_indices):
    df.iloc[:, column_indices] = df.iloc[:, column_indices].map(calculate_max)
    df = df.drop(columns=['RecoParticleTrack_absZ0sig'])
    for column in df.columns:
        if df[column].dtype == 'object':
            try:
                df[column] = pd.to_numeric(df[column], errors='coerce')
            except ValueError:
                print(f"Unable to convert column '{column}' to numeric.")
    return df
```
