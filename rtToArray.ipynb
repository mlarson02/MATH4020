```python
import uproot
import pandas as pd

def convert_root_to_dataframe(root_file_path, tree_name):
    # Open the ROOT file and get the TTree
    root_file = uproot.open(root_file_path)
    tree = root_file[tree_name]

    # Get the TBranches and their names
    branches = tree.arrays(library='np')
    branch_names = list(branches.keys())

    # Create a Pandas DataFrame
    dataframe = pd.DataFrame(branches)
    print(dataframe.size)
    return dataframe, branch_names


def save_to_csv(dataframe, csv_file_path):
    # Write Pandas DataFrame to CSV file
    dataframe.to_csv(csv_file_path, index=False)

# Example usage
#root_file_path = "../../../../../../../eos/user/m/mlarson/Reco_output_stage1_ML_IDEA_back/p8_ee_ZH_ecm240/chunk0.root"  # Replace with the path to your ROOT file
#root_file_path = "../../../../../../../eos/user/m/mlarson/Reco_output_stage1_ML_IDEA_back/p8_ee_ZH_ecm240/chunk0.root"
root_file_path = "ZZ_back.root"
tree_name = "events"  # Replace with the name of your TTree
resulting_numpy_array, branch_names = convert_root_to_dataframe(root_file_path, tree_name)

#csv_file_path = "../../../../../../../eos/user/m/mlarson/Reco_output_stage1_ML_IDEA_back/CSVOutput/ZH_back.csv"
csv_file_path = "csvs/ZZ_back.csv"
# Save NumPy array to CSV
save_to_csv(resulting_numpy_array, csv_file_path)

# Print the resulting NumPy array
print(resulting_numpy_array)
print(f"CSV file '{csv_file_path}' has been created.")
```
